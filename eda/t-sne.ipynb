```python
import os
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load real text data
real_text_path = "./data/datasets/bbc/politics"
real_texts = []
for filename in os.listdir(real_text_path):
    with open(os.path.join(real_text_path, filename), 'r') as f:
        real_texts.append(f.read())

# Load generated text data
folder_path = '../generated/cleaned/data_generated_gpt'
generated_texts = []
for filename in os.listdir(folder_path):
    with open(os.path.join(folder_path, filename), 'r') as f:
        generated_texts.append(f.read())

# Combine real and generated texts
all_texts = real_texts + generated_texts
labels = ['real'] * len(real_texts) + ['generated'] * len(generated_texts)

# Extract features using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(all_texts)

# Perform t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X.toarray())

# Plot the t-SNE visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:len(real_texts), 0], X_tsne[:len(real_texts), 1], label='Real', alpha=0.5)
plt.scatter(X_tsne[len(real_texts):, 0], X_tsne[len(real_texts):, 1], label='Generated', alpha=0.5)
plt.legend()
plt.title('t-SNE Visualization of Real and GPT Generated Text')
plt.show()
```
